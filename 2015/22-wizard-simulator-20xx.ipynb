```python
from collections import deque
```

Puzzle input

Hit Points: 71  
Damage: 10


```python
spells = {
    'missile': 53,
    'drain': 73,
    'shield': 113,
    'poison': 173,
    'recharge': 229,
}
```

```python
init = {
    'hp': 50,
    'mana': 500,
    'armor': 0,
    
    'spent': 0,
    
    'shield': 0,
    'poison': 0,
    'recharge': 0,
    
    'boss_hp': 71,
    'boss_damage': 10,
}

q = deque()
for spell in ['missile', 'drain', 'shield', 'poison', 'recharge']:
    state = init.copy()
    state['spell'] = spell
    q.append(state)
```

```python
def round(q=q):
    D = q.popleft()
    spell = D['spell']
    cost = spells[spell]
    if cost > D['mana']:
        return
    
    D['mana'] -= cost
    D['spent'] += cost
    
    if spell in ('shield', 'poison', 'recharge') and D[spell] > 1:
        return # illegal
    
    # wizard's turn
    
    if D['shield']:
        D['shield'] -= 1
        if D['shield'] == 0:
            D['armor'] -= 7
    
    if D['poison']:
        D['boss_hp'] -= 3
        D['poison'] -= 1
        
    if D['recharge']:
        D['mana'] += 101
        D['recharge'] -= 1
    
    # cast spell

    if spell == 'missile':
        D['boss_hp'] -= 4
    
    if spell == 'drain':
        D['boss_hp'] -= 2
        D['hp'] += 2
    
    if D['boss_hp'] <= 0:
        # wizard wins
        return D
        
    if spell == 'shield':
        D['armor'] += 7
        D['shield'] = 6
    
    if spell == 'poison':
        D['poison'] = 6
    
    if spell == 'recharge':
        D['recharge'] = 5
    
    # Boss turn
    
    if D['shield']:
        D['shield'] -= 1
        if D['shield'] == 0:
            D['armor'] -= 7
    
    if D['poison']:
        D['boss_hp'] -= 3
        D['poison'] -= 1
        
    if D['recharge']:
        D['mana'] += 101
        D['recharge'] -= 1

    if D['boss_hp'] <= 0:
        return D
    
    damage = max(1, D['boss_damage'] - D['armor'])
    
    D['hp'] -= damage
    
    if D['hp'] <= 0:
        return
    
    for nextspell in ['missile', 'drain', 'shield', 'poison', 'recharge']:
        E = D.copy()
        E['spell'] = nextspell
        q.append(E)
```

```python
for _ in range(1_000_000):
    D = round()
    if D:
        print(D)
        break
```
